```python
import tensorflow as tf
print(tf.__version__)
fmnist=tf.keras.datasets.fashion_mnist
(training_images,training_labels),(test_images,test_labels)=fmnist.load_data()
import numpy as np
import matplotlib.pyplot as plt
# choose any index between 0 to 60000
index = 377
#printing the complete array
np.set_printoptions(linewidth= 375)
print(f'LABEL{training_labels[index]}')
print(f'\nImage pixel array:\n{training_images[index]}')
#image using matplotlib
plt.imshow(training_images[index])
#normalizing the values of test images and training images
training_images= training_images/255.0
test_images= test_images/255.0
#building classification model
model=tf.keras.models.Sequential([tf.keras.layers.Flatten()
,tf.keras.layers.Dense(128,activation=tf.nn.relu),tf.keras.layers.Dense(10,activation=tf.nn.softmax)])
#softmax takes a list of values and scales these of all elements will be equal to 1.

#Declare sample inputs and convert to a tensor
inputs = np.array([[1.0, 3.0, 4.0, 2.0]])
inputs = tf.convert_to_tensor(inputs)
print(f'input to softmax function: {inputs.numpy()}')

#Feed the input to a softmax activation function
outputs = tf.keras.activations.softmax(inputs)
print(f'output of softmax function: {outputs.numpy()}')
#Get the sum of all values after the softmax
sum = tf.reduce_sum(outputs)
print(f'sum of the output: {sum}')

#get the index with highest index..argmax is the index of the maximum element in array
prediction= np.argmax(outputs)
print(f'class with highest probability: {prediction}')

model.compile(optimizer = tf.optimizers.Adam(),
              loss = 'sparse_categorical_crossentropy',
              metrics=['accuracy'])

model.fit(training_images, training_labels, epochs=50)
model.evaluate(test_images,test_labels)
```
